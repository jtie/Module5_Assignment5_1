### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes

    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination

    location of the user, destination, and the venue, and we mark the distance between each

    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM

3. Coupon attributes

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
df_bar = df[df['coupon'] == 'Bar'].copy()
```

2. What proportion of bar coupons were accepted?


```python
bar_coupon_accept_prop = df_bar['Y'].mean()
bar_coupon_accept_percentage = bar_coupon_accept_prop * 100

print(f"Bar Coupon Acceptance Proportion: {bar_coupon_accept_prop:.4f}")
print(f"Bar Coupon Acceptance Percentage: {bar_coupon_accept_percentage:.2f}%")
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# 1. Define value subsets
less_than_or_equal_to_3 = ['never', 'less1', '1~3']
more_than_3 = ['4~8', 'gt8']

# 2. Divide datasets based on the bar visiting frequency
less_than_or_equal_to_3_subset = df_bar[df_bar['Bar'].isin(less_than_or_equal_to_3)]
more_than_3_subset = df_bar[df_bar['Bar'].isin(more_than_3)]

# 3. Calculate subset acceptance rates
rate_less_than_or_equal_to_3_subset = less_than_or_equal_to_3_subset['Y'].mean() * 100
rate_more_than_3_subset = more_than_3_subset['Y'].mean() * 100

# 4. Print the difference of subset acceptance rates
print(f"Subset acceptance rate for drivers visiting bars ≤ 3 times/month: {rate_less_than_or_equal_to_3_subset:.2f}%")
print(f"Subset acceptance rate for drivers visiting bars > 3 times/month: {rate_more_than_3_subset:.2f}%")
print(f"Subset acceptance rate difference: {rate_more_than_3_subset - rate_less_than_or_equal_to_3_subset:.2f}% percentage points")
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# 1. Define value subsets
equal_to_or_more_than_1 = ['1~3', '4~8', 'gt8']
less_than_21_age = ['below21', '21']

# 2. Define the filter condition for target group: must visit > 1 time/mo AND age must NOT be 21 or below)
filter_condition = (
    df_bar['Bar'].isin(equal_to_or_more_than_1) & 
    ~df_bar['age'].isin(less_than_21_age)
)

# 3. Divide the data set to two subsets using filter condition
target_subset = df_bar[filter_condition]
all_others_outside_target_subset = df_bar[~filter_condition]

# 4. Calculate subset acceptance rates
rate_target = target_subset['Y'].mean() * 100
rate_others = all_others_outside_target_subset['Y'].mean() * 100

# 5. Print the difference of subset acceptance rates
print(f"Target subset acceptance rate (Bar > 1/mo AND Age > 25): {rate_target:.2f}%")
print(f"All others acceptance rate (All other drivers): {rate_others:.2f}%")
print(f"Subset acceptance rate difference: {rate_target - rate_others:.2f}% percentage points")
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
# 1. Define value subsets
equal_to_or_more_than_1 = ['1~3', '4~8', 'gt8']

# 2. Define the filter condition for target group: visits bar > 1/mo AND passenger is NOT 'Kid(s)' AND occupation is NOT farming/fishing/forestry)
filter_condition = (
    df_bar['Bar'].isin(equal_to_or_more_than_1) & 
    (df_bar['passanger'] != 'Kid(s)') & 
    (df_bar['occupation'] != 'Farming Fishing & Forestry')
)

# 3. Divide the data set to two subsets using filter condition
target_subset = df_bar[filter_condition]
all_others_outside_target_subset = df_bar[~filter_condition]

# 4. Calculate subset acceptance rates
rate_target = target_subset['Y'].mean() * 100
rate_others = all_others_outside_target_subset['Y'].mean() * 100

# 5. Print the difference of subset acceptance rates
print(f"Target subset acceptance rate (Bar > 1/mo, No Kid passengers, Non-Farming occupation): {rate_target:.2f}%")
print(f"All others acceptance rate (All other drivers): {rate_others:.2f}%")
print(f"Subset acceptance rate difference: {rate_target - rate_others:.2f}% percentage points")
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
# 1. Define value subsets
equal_to_or_more_than_1 = ['1~3', '4~8', 'gt8']
less_than_30_age = ['below21', '21', '26']
visits_cheap_resturant = ['4~8', 'gt8']
income_under_50k = ['Less rather than $12500', '$12500 - $24999', '$25000 - $37499', '$37500 - $49999']

# 2. Define Filter Condition 1: Bar > 1/mo AND passenger not kid AND marital status not widowed
filter_condition_1 = (
    df_bar['Bar'].isin(equal_to_or_more_than_1) & 
    (df_bar['passanger'] != 'Kid(s)') & 
    (df_bar['maritalStatus'] != 'Widowed')
)

# 3. Define Filter Condition 2: Bar > 1/mo AND age < 30
filter_condition_2 = (
    df_bar['Bar'].isin(equal_to_or_more_than_1) & 
    df_bar['age'].isin(less_than_30_age)
)

# 4. Define Filter Condition 3: Cheap Restaurant > 4/mo AND income < 50K
filter_condition_3 = (
    df_bar['RestaurantLessThan20'].isin(visits_cheap_resturant) & 
    df_bar['income'].isin(income_under_50k)
)

# 5. Calculate subset acceptance rates
subset_rate_1 = df_bar[filter_condition_1]['Y'].mean() * 100
subset_rate_2 = df_bar[filter_condition_2]['Y'].mean() * 100
subset_rate_3 = df_bar[filter_condition_3]['Y'].mean() * 100

# 6. Print the difference of subset acceptance rates
print("--- Coupon Acceptance Rate Comparison ---")
print(f"Subset 1 (Bar >1/mo, No Kids, Not Widowed): {subset_rate_1:.2f}%")
print(f"Subset 2 (Bar >1/mo, Age < 30):             {subset_rate_2:.2f}%")
print(f"Subset 3 (Cheap Rest >4/mo, Income < 50K):  {subset_rate_3:.2f}%")
```

```python
#2. Define the filter condition for subset1: CoffeeHouse visits equal to or more than 1 AND drive with passanger 'Friend(s)' or 'Partner'
equal_to_or_more_than_1 = ['1~3', '4~8', 'gt8']
filter_condition1 = (
    df_coffee['CoffeeHouse'].isin(equal_to_or_more_than_1) &
    df_coffee['passanger'].isin(['Friend(s)', 'Partner'])
)

# 3. Define the data subset1 to two subsets using filter condition
target_subset1 = df_coffee[filter_condition1]

# 4. Calculate subset1 acceptance rates
rate_subset1 = target_subset1['Y'].mean() * 100

# 5. Print subset1 acceptance rates
print(f"Total rows of subset1 where CoffeeHouse visits equal to or more than 1 AND passenger equals 'Friend(s)' or 'Partner': {len(target_subset1)}")
print(f"Target Driver Subset1 Acceptance Rate: {rate_subset1:.2f}%")
```

```python
#6. Define the filter condition for subset2: CoffeeHouse visits equal to or more than 1 AND drive with passanger 'Alone' or 'Kid(s)'
equal_to_or_more_than_1 = ['1~3', '4~8', 'gt8']
filter_condition2 = (
    df_coffee['CoffeeHouse'].isin(equal_to_or_more_than_1) &
    df_coffee['passanger'].isin(['Alone', 'Kid(s)'])
)

# 7. Define the data subset2 to two subsets using filter condition
target_subset2 = df_coffee[filter_condition2]

# 8. Calculate subset2 acceptance rates
rate_subset2 = target_subset2['Y'].mean() * 100

# 9. Print subset2 acceptance rates
print(f"Total rows of subset2 where CoffeeHouse visits equal to or more than 1 AND passenger equals 'Alone' or 'Kid(s)': {len(target_subset2)}")
print(f"Target Driver Subset2 Acceptance Rate: {rate_subset2:.2f}%")
```

```python
# 10. Create two new data frames for two condition means
acceptance_rate1 = df_coffee[filter_condition1]['Y'].mean()
acceptance_rate2 = df_coffee[filter_condition2]['Y'].mean()

# 11. Create a temporary DataFrame for plotting
pd_data = pd.DataFrame(
    {
        "Passenger Group": ["Friends / Partner", "Alone / Kids"],
        "Acceptance Rate": [acceptance_rate1, acceptance_rate2],
    }
)

# 12. Generate the bar plot
plt.figure(figsize=(6, 4))
ax = sns.barplot(
    x="Passenger Group",
    y="Acceptance Rate",
    data=pd_data,
    palette="Set2",
    hue="Passenger Group",
    legend=False,
)

# 13. Format the Y-axis as percentage and set limits
plt.ylim(0, 1)
plt.title("Coffee House Coupon Acceptance Rate By Passenger Groups")
plt.xlabel("Passenger Group")
plt.ylabel("Acceptance Rate")

# 14. Add percentage text on top of the bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.1%}",
        (p.get_x() + p.get_width() / 2.0, p.get_height()),
        ha="center",
        va="center",
        xytext=(0, 8),
        textcoords="offset points",
    )

# 15. Display the plot
plt.savefig("images/coffee_house_coupon_acceptance_rate.png", dpi=300, bbox_inches='tight')
plt.close()
```
